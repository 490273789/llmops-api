```python
from langchain_core.output_parsers import StrOutputParser
from langchain.chat_models import init_chat_model
from langchain_core.prompts import ChatPromptTemplate

llm = init_chat_model(model="deepseek-chat", model_provider="deepseek")

prompt = ChatPromptTemplate.from_template("{query}")

chain = prompt | llm
res = chain.invoke({"query": "hello"})

print(f"res: {res}")
```

```python
parser = StrOutputParser()
output = parser.invoke(res)
print(type(output))
print(output)
```

JsonOutPutParse

```python
from langchain_core.pydantic_v1 import BaseModel, Field
from langchain_core.output_parsers import JsonOutputParser


class Joke(BaseModel):
    joke: str = Field(description="回答用户问题")
    punchline: str = Field(description="这个冷笑话的笑点")


parser = JsonOutputParser(pydantic_object=Joke)

prompt = ChatPromptTemplate.from_template(
    "请根据用户的提问进行回答。\n{format_instruction}\n{query}"
).partial(format_instruction=parser.get_format_instructions)

chain = prompt | llm | parser
joke = chain.invoke({"query": "请讲一个程序员的冷笑话"})

print(joke)
```
